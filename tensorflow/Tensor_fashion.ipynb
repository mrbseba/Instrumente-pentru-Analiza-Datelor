```python
import tensorflow as tf
```

```python
# Incarcam setul de date Fashion MNIST
fmnist = tf.keras.datasets.fashion_mnist
```

```python
# Incarca subseturile de training și test din Fashion MNIST
(training_images,training_labels), (test_images, test_labels) = fmnist.load_data()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Puteti indica valori intre 0 si 59999
index = 0
# Setam numarul de caractere pe rand la tipar
np.set_printoptions(linewidth=320)
# Tiparim eticheta și matricea imaginii
print(f'LABEL: {training_labels[index]}')
print(f'\nIMAGE PIXEL ARRAY:\n {training_images[index]}')
# Visualizam imaginea
plt.imshow(training_images[index])
```

```python
# Normalizare valorilor numerice a fiecărui pixel

training_images = training_images / 255.0
test_images = test_images / 255.0
```

```python
# Build the classification model
model = tf.keras.models.Sequential(
[tf.keras.layers.Flatten(),
tf.keras.layers.Dense(128,activation=tf.nn.relu),
tf.keras.layers.Dense(10,activation=tf.nn.softmax)])
```

```python
model.compile(optimizer = tf.optimizers.Adam(),
loss = 'sparse_categorical_crossentropy',
metrics=['accuracy'])
model.fit(training_images, training_labels, epochs=5)
```

```python
# Evaluate the model on unseen data
model.evaluate(test_images, test_labels)
```
