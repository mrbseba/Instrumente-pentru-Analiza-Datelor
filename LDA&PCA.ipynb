```python
# LDA data catergorizate - presupune ca datele sunt distribuite normal
from sklearn.datasets import load_wine

wine = load_wine()

X = wine.data
y = wine.target

print('Wine dataset: ', X.shape)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

# pentru LDA nu ar trebuie standatizare - da spoate de facut
X_scaled = StandardScaler().fit_transform(X)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis(n_components=2)

X_lda = lda.fit_transform(X_scaled, y)
```

```python
# vizualizarea distribuirea datelor
import matplotlib.pyplot as plt 

plt.figure(figsize=[7, 5])

plt.scatter(X_lda[:, 0], X_lda[:, 1], c = y, s = 25, cmap = 'plasma')

plt.title("LDA for wine data with 2 components")
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

iris = load_iris()

X = iris.data
y = iris.target

# normalize the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaled = sc.fit_transform(X)

# create PCA model
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# create LDA model
lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X, y)

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(6, 12))

sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette='Set1', ax=ax[0])
sns.scatterplot(x=X_lda[:, 0], y=X_lda[:, 1], hue=y, palette='Set1', ax=ax[1])

ax[0].set_title('PCA of IRIS dataset', fontsize=15, pad=15)
ax[1].set_title('LDA of IRIS dataset', fontsize=15, pad=15)

ax[0].set_xlabel('PC1', fontsize=12)
ax[1].set_xlabel('LD1', fontsize=12)

ax[0].set_ylabel('PC2', fontsize=12)
ax[1].set_ylabel('LD2', fontsize=12)

plt.show()
```
